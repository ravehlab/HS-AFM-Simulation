```python
# import dependencies
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
# load data
# f_path = "../../HS-AFM-Dataset/dataset/afm/new/mini_0uM_100ns_40x40/20.pickle"
# f_path = "../results_10ns_1010.pickle"
# f_path = "../results_1ns_4040.pickle"
f_path = "../results_1ns_8080.pickle"


with open(f_path, 'rb') as f:
    afm_data = pickle.load(f)

non_rasterized_maps = np.dstack(afm_data["non_rasterized_maps"])
```

```python
# process data
averaged_map = np.mean(non_rasterized_maps, axis=2) - 40
fig, ax = plt.subplots()
im = ax.imshow(averaged_map, cmap="jet", vmin = 0, vmax=30)
cbar = fig.colorbar(im, ax=ax)
cbar.set_label("Height (nm)")
plt.show()
```
